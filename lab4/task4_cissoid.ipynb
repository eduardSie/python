Намалювати криву та супутні елементи.

**Рівняння:**  
$$y^2 = \frac{x^3}{2a - x}$$

**Параметричне задання:**  
$$x = 2a\sin^2 t, \quad y = 2a\sin^2 t \cdot \tan t = \frac{2a\sin^3 t}{\cos t}$$

**Супутні елементи:**  
- Кола діаметром $2a$ (з якого будується цисоїда)  
- Асимптота $x = 2a$  
- Допоміжні лінії побудови  

Використання циклів заборонено.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 2.0

t = np.linspace(-1.45, 1.45, 2000)
x_c = 2*a * np.sin(t)**2
y_c = 2*a * np.sin(t)**2 * np.tan(t)

mask = np.abs(x_c - 2*a) > 0.01
x_c, y_c = x_c[mask], y_c[mask]

theta = np.linspace(0, 2*np.pi, 500)
x_circ = a + a * np.cos(theta)
y_circ = a * np.sin(theta)

t_samp = np.array([-1.1, -0.7, 0.7, 1.1])
xs_samp = 2*a * np.sin(t_samp)**2
ys_samp = 2*a * np.sin(t_samp)**2 * np.tan(t_samp)
xb_samp = a + a * np.cos(2*t_samp)
yb_samp = a * np.sin(2*t_samp)

fig, ax = plt.subplots(figsize=(10, 9))

ax.plot(x_circ, y_circ, color='#888888', linewidth=1.2, linestyle='--',
        label=f'Кола діаметром $2a$ (a={a})')

ax.axvline(x=2*a, color='#FF6B35', linewidth=1.5, linestyle=':',
           label=f'Асимптота: $x = 2a = {2*a}$')

ax.plot(x_c, y_c, color='#1565C0', linewidth=2.5,
        label=r'Цисоїда Діокла: $y^2 = \dfrac{x^3}{2a-x}$')

ax.plot(np.vstack([np.zeros(len(t_samp)), xb_samp]),
        np.vstack([np.zeros(len(t_samp)), yb_samp]),
        color='#4CAF50', linewidth=0.8, alpha=0.6, linestyle='-.')
ax.plot(xb_samp, yb_samp, 'o', color='#4CAF50', markersize=5, alpha=0.7,
        label='Точки на колі (побудова)')
ax.plot(xs_samp, ys_samp, 'o', color='#1565C0', markersize=5)

ax.plot(0, 0, 'ko', markersize=7, zorder=5)
ax.annotate('O', (0, 0), (-0.3, -0.5), fontsize=10)
ax.plot(2*a, 0, 'rs', markersize=8, zorder=5)
ax.annotate(f'A(2a, 0)', (2*a, 0), (2*a+0.1, 0.3), fontsize=9, color='red')

ax.text(0.5, 4.5, r'$y^2 = \dfrac{x^3}{2a - x}$', fontsize=14,
        color='#1565C0', fontweight='bold')
ax.text(0.5, 3.5, f'Параметр $a = {a}$', fontsize=11, color='#555')

ax.set_xlim(-0.5, 2*a + 1.2); ax.set_ylim(-6, 6)
ax.set_aspect('equal')
ax.set_xlabel('$x$', fontsize=13); ax.set_ylabel('$y$', fontsize=13)
ax.set_title('Сєров Едуард, К-26 | ЛР№4 | Задача 4 (Варіант 11)\nЦисоїда Діокла з супутніми елементами',
             fontsize=13, fontweight='bold')
ax.legend(fontsize=10, loc='upper left')
ax.grid(True, alpha=0.25)
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

plt.tight_layout()
plt.savefig('task4_cissoid.png', dpi=150, bbox_inches='tight')
plt.show()
```
