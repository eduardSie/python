Намалювати тривимірну поверхню як wireframe plot.

**Рівняння:**  
$$\frac{x^2}{a^2} + \frac{y^2}{b^2} - \frac{z^2}{c^2} = 1$$

**Параметричне задання:**  
$$x = a \cos u \cosh v, \quad y = b \sin u \cosh v, \quad z = c \sinh v$$
де $u \in [0, 2\pi]$, $v \in \mathbb{R}$.

Використання циклів заборонено.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.lines import Line2D
```

```python
a_p, b_p, c_p = 2.0, 1.5, 1.0

u = np.linspace(0, 2*np.pi, 60)
v = np.linspace(-1.5, 1.5, 40)
U, V = np.meshgrid(u, v)

X = a_p * np.cos(U) * np.cosh(V)
Y = b_p * np.sin(U) * np.cosh(V)
Z = c_p * np.sinh(V)

fig = plt.figure(figsize=(10, 9))
ax = fig.add_subplot(111, projection='3d')

ax.plot_wireframe(X, Y, Z, color='#1565C0', linewidth=0.5, alpha=0.7, rstride=1, cstride=2)

ax.set_xlabel('$x$', fontsize=12); ax.set_ylabel('$y$', fontsize=12); ax.set_zlabel('$z$', fontsize=12)
ax.set_title('Сєров Едуард, К-26 | ЛР№4 | Задача 5 (Варіант 3)\nОднопорожнинний гіперболоїд',
             fontsize=12, fontweight='bold')

eq_label = r'$\dfrac{x^2}{a^2} + \dfrac{y^2}{b^2} - \dfrac{z^2}{c^2} = 1$' +            f'\n$a={a_p},\\ b={b_p},\\ c={c_p}$'
legend_el = [Line2D([0],[0], color='#1565C0', linewidth=1.5, label=eq_label)]
ax.legend(handles=legend_el, loc='upper left', fontsize=11, handlelength=1.5, framealpha=0.9)

ax.view_init(elev=25, azim=45)
plt.tight_layout()
plt.savefig('task5_hyperboloid.png', dpi=150, bbox_inches='tight')
plt.show()
```
