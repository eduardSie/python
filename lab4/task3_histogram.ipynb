За даними задачі 4 ЛР№2 побудувати гістограму, яка відображає розподіл точок по інтервалах.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('lab2_var090.csv')
n_raw = df['n'].values
print(f'Кількість точок: {len(n_raw)}, діапазон: [{n_raw.min()}, {n_raw.max()}]')
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
n_bins = 8
counts, edges, patches = ax.hist(n_raw, bins=n_bins, color='steelblue',
                                  edgecolor='white', linewidth=0.8, alpha=0.85, rwidth=0.88)

ax.set_xlabel('Розмір вхідних даних $n$', fontsize=12)
ax.set_ylabel('Кількість точок', fontsize=12)
ax.set_title('Сєров Едуард, К-26 | ЛР№4 | Задача 3\nРозподіл точок за інтервалами розміру вхідних даних',
             fontsize=12, fontweight='bold')
ax.grid(True, alpha=0.3, axis='y')

for patch in ax.patches:
    h = patch.get_height()
    if h > 0:
        ax.annotate(f'{int(h)}', xy=(patch.get_x() + patch.get_width()/2, h),
                    xytext=(0, 4), textcoords='offset points', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.savefig('task3_histogram.png', dpi=150, bbox_inches='tight')
plt.show()
```
